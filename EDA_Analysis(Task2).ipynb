# EDA - Uni and Bi Variant Analysis for dataset1


--> Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that involves examining and visualizing the characteristics of a dataset.

**Univariate Analysis - Numeric Variables:**

-->  Univariant analysis focuses on exploring the distribution and characteristics of a single variable.

Histogram is used to plot the graphical representation of a numerical variable. The x-axis represents the numeric data i.e age and y-axis represents the respective count of the data.

```python
# Histogram for a numeric variable
plt.figure(figsize=(10, 6))
sns.histplot(dataset1['age'], bins=30, kde=True)
plt.title('Histogram of Numeric Variable')
plt.show()
```

Box plot is created to get the summary of variable ie. minimum, first quartile, median, 3rd quartile and maximum.

```python
# Boxplot for a numeric variable
plt.figure(figsize=(10, 6))
sns.boxplot(x=dataset1['balance'])
plt.title('Boxplot of Balance')
plt.show()
```

Univariate Analysis - Categorical Variables:



```python
# Countplot for a categorical variable
plt.figure(figsize=(20, 12))
sns.countplot(x='country', data=dataset1)
plt.title('Countplot of Country')
plt.show()
```

Univariate Analysis - Discrete Variables:



--> To find the relationship between a numeric and categorical variable we use Barplot

```python
# Barplot for a discrete variable
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='balance', data=dataset1, estimator=sum)
plt.title('Barplot of Discrete Variable(age) vs. Numeric Variable(balance)')
plt.show()
```

--> Bivariate analysis involves exploring the relationship between two variables.

--> Scatterplot uses dots to represent the relationship between two numeric variables.

```python
# Scatter plot for two numeric variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='credit_score', data=dataset1)
plt.title('Scatter Plot of age vs. credit_score')
plt.show()
```

**Multivariate Analysis**

--> Multivariate means the analysis of more than two variables.



--> Pair plot plots the pairwise relationship in a dataset.

```python
# Pairplot for multiple numeric variables
sns.pairplot(dataset1[['country', 'balance', 'estimated_salary']])
plt.suptitle('Pairplot of Numeric Variables')
plt.show()
```

--> Heatmap is used to visualise the correlations between pairs of numeric variables.


```python
# Correlation matrix for numeric variables
correlation_matrix = dataset1[['customer_id', 'age', 'churn']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
